```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020 = pd.read_csv("Resources/athletic_sales_2020.csv",low_memory=False)
athletic_sales_2021 = pd.read_csv("Resources/athletic_sales_2021.csv",low_memory=False)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
frames = [athletic_sales_2020,athletic_sales_2021]
athletic_sales = pd.concat(frames)

athletic_sales.reset_index()
athletic_sales.head(10)
# print(athletic_sales.info())
# print("")
# print(len(athletic_sales))

# print(8346+1297)
```

```python
# Check if any values are null.
# since it comes out false
athletic_sales.isnull().values.any()
# since the value printed is False, there ar no null values across the entire dataframe.
```

```python
# Check the data type of each column
athletic_sales.info()
```

```python
# Convert the "invoice_date" to a datetime datatype

athletic_sales["invoice_date"] = pd.to_datetime(athletic_sales["invoice_date"],format="%m%d%YY")

# athletic_sales.head(30)

```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales.info()
# column type is now datetime. convertion was completed.
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
# sales_totals2 = athletic_sales.groupby("region","state","city")
sales_totals = athletic_sales.groupby(["region","state","city"]).agg(Total_Products_Sold=("units_sold","sum"))
sales_totals.sort_values(by="Total_Products_Sold",ascending=False, inplace=True)

# Show the top 5 results.
sales_totals.head(5)
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
sales_pivot = athletic_sales.pivot_table(index=["region","state","city"],values="units_sold",aggfunc="sum")
# Rename the "units_sold" column to "Total_Products_Sold"
sales_pivot.rename(columns={"units_sold":"Total_Products_Sold"}, inplace=True)

# Show the top 5 results.
sales_pivot.sort_values("Total_Products_Sold",ascending=False,inplace=True)
sales_pivot.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_dollar_GB = athletic_sales.groupby(["region","state","city"]).agg(Total_Sales=("total_sales","sum"))
sales_dollar_GB.sort_values(by="Total_Sales",ascending=False, inplace=True)
# Rename the "total_sales" column to "Total Sales"
sales_dollar_GB.rename(columns={"Total_Sales":"Total Sales"}, inplace=True)

# Show the top 5 results.
sales_dollar_GB.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_dollar_pivot = athletic_sales.pivot_table(index=["region","state","city"],values="total_sales",aggfunc="sum")

# Optional: Rename the "total_sales" column to "Total Sales"
sales_dollar_pivot.rename(columns={"total_sales":"Total Sales"}, inplace=True)
sales_dollar_pivot.sort_values(by="Total Sales",ascending=False, inplace=True)

# Show the top 5 results.
sales_dollar_pivot.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
print(athletic_sales["product"].unique())
# just checking the spelling and consistency for "Women's Athletic Footwear". Looks consistent.

print("len of all: ",len(athletic_sales))
# total of 9643 records in total data frame.
womens_ath_footwear = athletic_sales.loc[athletic_sales["product"] == "Women's Athletic Footwear"]
print("")
print("len of just womens ath shoes: ",len(womens_ath_footwear))
# the records were product is women's athletic footwear are 1604, so sanity check passes.
womens_ath_footwear.info()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_ath_footwear_sum = womens_ath_footwear.groupby(["retailer","region", "state","city"]).agg(Womens_Footwear_Units_Sold=("units_sold","sum"))
womens_ath_footwear_sum.sort_values(by="Womens_Footwear_Units_Sold",ascending=False, inplace=True)


# Show the top 5 results.
womens_ath_footwear_sum.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_ath_footwear_PVT = womens_ath_footwear.pivot_table(index=["retailer","region", "state","city"],values="units_sold",aggfunc="sum")
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_ath_footwear_PVT.rename(columns={"units_sold":"Womens_Footwear_Units_Sold"}, inplace=True)
womens_ath_footwear_PVT.sort_values(by="Womens_Footwear_Units_Sold",ascending=False, inplace=True)
# Show the top 5 results.
womens_ath_footwear_PVT.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_ath_footwear_by_day = womens_ath_footwear.pivot_table(index="invoice_date",values="total_sales",aggfunc="sum")
womens_ath_footwear_by_day.sort_values("total_sales",ascending=False,inplace=True)

# Optional: Rename the "total_sales" column to "Total Sales"
womens_ath_footwear_by_day.rename(columns={"total_sales":"Total Sales"},inplace=True)
# Show the table.
womens_ath_footwear_by_day.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
womens_ath_fwear_by_day_resamp = womens_ath_footwear_by_day.resample("D").sum()
womens_ath_fwear_by_day_resamp.sort_values("Total Sales",ascending=False,inplace=True)
# Sort the resampled pivot table in ascending order on "Total Sales".
womens_ath_fwear_by_day_resamp.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
womens_ath_fwear_by_week_resamp = womens_ath_footwear_by_day.resample("W").sum()
# print("num records for Daily womens_ath_footwear_by_day is: ",len(womens_ath_footwear_by_day))
# print("num records for WEEKLY womens_ath_fwear_by_week_resamp is: ",len(womens_ath_fwear_by_week_resamp))

# Sort the resampled pivot table in ascending order on "Total Sales".
womens_ath_fwear_by_week_resamp.sort_values("Total Sales",ascending=False,inplace=True)
womens_ath_fwear_by_week_resamp.head(10)
```
